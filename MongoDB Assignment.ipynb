```python
# Ans 1:
'''
MongoDB is a popular NoSQL (non-relational) database management system that stores data in JSON-like documents with dynamic schema. 
It is designed to be scalable, flexible, and high-performing.

Non-relational databases, also known as NoSQL databases, differ from traditional relational databases (SQL databases) in the way they store and retrieve data. 
Unlike SQL databases, NoSQL databases don't use a fixed schema to define the structure of the data. 
Instead, they use flexible schemas that can be easily adapted to changing data requirements. 
NoSQL databases also tend to be horizontally scalable, meaning that they can handle large amounts of data by adding more servers to a cluster.

In scenarios where the data requirements are changing frequently, NoSQL databases like MongoDB are often preferred over SQL databases. 
NoSQL databases can easily adapt to these changes without the need for complex data migrations or schema changes. 
They can also handle large amounts of unstructured or semi-structured data, such as text, images, and videos, that may not fit well in a rigidly structured SQL database.

MongoDB is particularly well-suited for use cases such as:

1. Storing large volumes of data with high write throughput and low latency requirements, such as IoT data or real-time analytics data.

2. Building applications that require a flexible data model that can easily accommodate changes to the data structure over time.
3. Creating geospatial applications that can take advantage of MongoDB's built-in support for geospatial indexing and querying.

4. Implementing microservices architectures where each microservice requires its own database instance, as MongoDB allows for easy horizontal scaling and distribution of data across multiple servers.

In summary, MongoDB is a NoSQL database that is often preferred over SQL databases in scenarios where the data requirements are changing frequently or where large amounts of unstructured data need to be stored and queried.
'''
```

```python
# # Ans 2:
'''
Some of the key features of MongoDB are:
'''
# 1. Document-oriented: MongoDB stores data as documents, which are similar to JSON objects. Each document can have a flexible schema, allowing for easy changes to the data structure.

# 2. Scalability: MongoDB is designed to be horizontally scalable, meaning that it can handle large amounts of data by adding more servers to a cluster. It also supports sharding, which allows data to be partitioned across multiple servers.

# 3. High availability: MongoDB supports replica sets, which are groups of MongoDB servers that maintain identical copies of data. If one server fails, another server in the replica set can take over, ensuring that data remains available.

# 4. Flexible indexing: MongoDB supports a variety of indexing options, including single-field, multi-field, and geospatial indexing. This makes it easy to query and search data in different ways.

# 5. Ad hoc queries: MongoDB supports ad hoc queries, which means that you can query the data in any way you want without having to define a schema or structure beforehand.

# 6. Aggregation: MongoDB provides powerful aggregation capabilities, allowing you to group, filter, and transform data in a variety of ways.

# 7. GridFS: MongoDB provides GridFS, a specification for storing and retrieving large files, such as images and videos, in a MongoDB database.

# 8. Integration: MongoDB integrates with many popular programming languages and frameworks, including Python, Java, and Node.js.
```

```python
# Ans 7:

'''
In MongoDB, delete_one(), delete_many(), and drop() are used to remove data from a MongoDB database.

>> delete_one() is used to remove a single document that matches a specific query. 
If multiple documents match the query, only the first one that is found is deleted.
The method takes a filter as its argument, which specifies the query criteria.

>> delete_many() is used to remove all documents that match a specific query. 
The method takes a filter as its argument, which specifies the query criteria.

>> drop() is used to remove an entire collection from a MongoDB database. 
The method takes no arguments and simply drops the entire collection.


# Here are some use cases for each method:

1. delete_one() can be used to delete a single document from a collection, such as a user account that has been deactivated.

2. delete_many() can be used to remove all documents that meet certain criteria, such as all orders placed by a specific customer.

3. drop() can be used to completely remove a collection from the database, such as when a collection is no longer needed.

It's important to note that these methods can have significant consequences on the data in a MongoDB database, 
so they should be used carefully and with caution. 
Always make sure to back up your data before performing any deletes or drops.
'''

# for example:

# Removing a single document with delete_one()
coll_pwskills.delete_one({'name': 'Suharsh'})

```

```python
# Removing multiple documents with delete_many()
coll_pwskills.delete_many({'id': {'$gte': 10}})

```
